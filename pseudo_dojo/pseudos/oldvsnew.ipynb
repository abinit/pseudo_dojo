```python
new_pseudos = DojoTable.from_dir("/Users/gmatteo/git_repos/pseudo_dojo/pseudo_dojo/pseudos/ONCVPSP-PBE-PDv0.4/")
```

## dataframe with new pseudos (new optimizer, same input file as in PDv0.3) 

```python
new_frame, errors = new_pseudos.get_dojo_dataframe()
new_frame["oncvpsp"] = "new"
new_frame["name"] = new_frame.index
new_frame[keys]
```

```python
new_frame[keys].describe()
```

## Results obtained with PDv0.3 (old optimizer)

```python
# Select old pseudos with a corresponding entry in the new_frame
old_frame, errors = old_pseudos.get_dojo_dataframe()
old_frame["oncvpsp"] = "old"
old_frame["name"] = old_frame.index
old_frame = old_frame.ix[new_frame.index]
old_frame[keys]
```

```python
old_frame[keys].describe()
```

## Difference between the results obtained with old and new pseudos 

```python
new_minus_old = new_frame[keys] - old_frame[keys]
new_minus_old
```

```python
# Take absolute value of errors and compute stats
abserr = new_minus_old.copy()
for k in abserr.keys():
    abserr[k] = abserr[k].abs()
abserr.describe()
```

## Pseudos for which the absolute change in the deltafactor in > 0.4

```python
new_minus_old[new_minus_old["high_dfact_meV"].abs() > 0.4]
```

## Pseudos for which the absolute change in the relative error of the GBRV fcc lattice in > 0.1

```python
new_minus_old[new_minus_old["high_gbrv_fcc_a0_rel_err"].abs() > 0.1]
```

## Pseudos for which the absolute change in the relative error of the GBRV bcc lattice in > 0.1

```python
new_minus_old[new_minus_old["high_gbrv_bcc_a0_rel_err"].abs() > 0.1]
```

## Plot results

```python
# Concatenate and sort by Z to prepare seaborn calls
data = pd.concat([old_frame, new_frame])
data = data.sort_values("Z");
```

```python
# Draw barplot with the deltfactor computed at the highest cutoff-energy
sns.factorplot(y="name", x="high_dfact_meV", hue="oncvpsp", data=data,
               size=24, kind="bar", palette="muted", orient="h", aspect=0.4)
```

```python
# Draw barplot with the relative error in the GBRV FCC benchmark
g = sns.factorplot(y="name", x="high_gbrv_fcc_a0_rel_err", hue="oncvpsp", data=data,
                   size=24, kind="bar", palette="muted", orient="h", aspect=0.4)
```

```python
# Draw barplot with the relative error in the GBRV BCC benchmark
g = sns.factorplot(y="name", x="high_gbrv_bcc_a0_rel_err", hue="oncvpsp", data=data,
                   size=24, kind="bar", palette="muted", orient="h", aspect=0.4)
```
